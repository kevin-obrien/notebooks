```python
import erddapy

# Define the ERDDAP server URL and dataset
server_url = 'https://data.pmel.noaa.gov/generic/erddap/'
dataset_id = 'cchdo_bottle'

# Create a connection to the ERDDAP server
e = erddapy.ERDDAP(server=server_url)

# Set the dataset ID for the query
e.dataset_id = dataset_id

def get_metadata():
    try:
        # Retrieve metadata from the dataset (in JSON format by default)
        metadata = e.metadata()

        # Print some basic metadata information
        print("Dataset Metadata:")
        print(f"Dataset ID: {metadata['dataset_id']}")
        print(f"Title: {metadata['title']}")
        print(f"Summary: {metadata['summary']}")
        print(f"Variable names: {metadata['variables']}")

        # Print out the metadata for all variables
        print("\nVariables:")
        for var in metadata['variables']:
            print(f"- {var}")

        # You can also access specific metadata fields
        print("\nSpecific variable metadata:")
        for var in metadata['variables']:
            var_metadata = e.variable_metadata(var)
            print(f"Metadata for {var}:")
            print(var_metadata)

    except Exception as e:
        print(f"Error retrieving metadata: {e}")

def main():
    get_metadata()

if __name__ == '__main__':
    main()
```
